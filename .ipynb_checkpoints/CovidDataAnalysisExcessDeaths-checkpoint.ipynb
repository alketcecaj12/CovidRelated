```python
import pandas as pd
import matplotlib.pyplot as plt

import glob
import pandas as pd

# get data file names
path =r'dati-andamento-nazionale'
filenames = glob.glob(path + "/*.csv")

dfs = []
for filename in filenames:
    dfs.append(pd.read_csv(filename))

# Concatenate all data into one DataFrame
big_frame = pd.concat(dfs, ignore_index=True)
```

```python
big_frame.to_csv('all_data.csv')
```

```python
big_frame.data = pd.to_datetime(big_frame['data'])
```

```python
d = big_frame.deceduti
plt.figure(figsize = (20, 4))
plt.plot(d)
plt.show()
```

```python
dati_dec = big_frame[['data', 'deceduti']]
```

```python
dati_dec.sort_values(by= ['data'], inplace = True)
```

```python
dati_dec
```

```python
dati_dec = dati_dec.drop_duplicates()
```

```python
dati_dec.head(50)
```

```python
dati_dec_gg = dati_dec.deceduti.diff()
```

```python
dati_dec['dec_gg'] = dati_dec_gg
```

```python
dati_dec.fillna(7, inplace = True)
```

```python
dati_dec
```

```python
plt.figure(figsize = (20, 4))
plt.plot(dati_dec.dec_gg)
plt.show()
```

```python
dati_dec.set_index('data', inplace = True)
```

```python
plt.figure(figsize = (20, 4))
plt.plot(dati_dec.dec_gg)
plt.show()
```

```python
dati_dec_2020 = dati_dec[:312]
dati_dec_2021 = dati_dec[312:]
```
